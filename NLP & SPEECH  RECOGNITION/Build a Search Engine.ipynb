```python
import nltk
from gtts import gTTS
from selenium import webdriver

import os
import speech_recognition as sr

import wolframalpha
import playsound
import pyaudio
num = 0

import smtplib
```

# ListenToUser()
#### This  function gets the audio from user using microphone

```python
def ListenToUser():
    r = sr.Recognizer()
    audio = ''
    with sr.Microphone() as source:
        print("Speak...")
        audio = r.listen(source, phrase_time_limit=5)
    print("Stop.")
    try:
        text = r.recognize_google(audio,language='en-US')
        print("You : ", text)
        return text
    except:
        Speak("Could not understand your audio, PLease try again!")
        
        return 0
```

# Speak() 
#### This function  converts the text to voice
#### The text in this context is generated by the Asistant



```python
def Speak(audio):
    global num
    num +=1
    print("You : ", audio)
    toSpeak = gTTS(text=audio, lang='en-US', slow=False)
    file = str(num)+".mp3"
    toSpeak.save(file)
    playsound.playsound(file, True)
    os.remove(file)
```

# WebSearch()
#### This lanches various smaller functions to intiate web search appropriate to the user command.

```python
def WebSearch(input):
    driver = webdriver.Firefox()
    driver.implicitly_wait(1)
    driver.maximize_window()
    
    if 'youtube' in input.lower():
        Speak("Opening in youtube")
        indx = input.lower().split().index('youtube')
        query = input.split()[indx+1:]
        driver.get("http://www.youtube.com/results?search_query=" + '+'.join(query))
        return

    elif 'wikipedia' in input.lower():
        Speak("Opening Wikipedia")
        indx = input.lower().split().index('wikipedia')
        query = input.split()[indx + 1:]
        driver.get("https://en.wikipedia.org/wiki/" + '_'.join(query))
        return
    else:
        if 'google' in input:
            Speak("Opening Google")
            indx = input.lower().split().index('google')
            query = input.split()[indx + 1:]
            driver.get("https://www.google.com/search?q=" + '+'.join(query))
        elif 'search' in input:
            Speak("Opening Google")
            indx = input.lower().split().index('google')
            query = input.split()[indx + 1:]
            driver.get("https://www.google.com/search?q=" + '+'.join(query))
        else:
            driver.get("https://www.google.com/search?q=" + '+'.join(input.split()))
        return
```

# Call_My_Assistant()
####  This is the main function which calls all other fnctions.

```python
def Call_My_Assistant():
    intro = '''Hello, I am Laila and Your personal Assistant at your service.'''
    Speak(intro)
    try:
        Speak("I could assist you in web search while you are busy with your work. For example, you will have to say such as ,  you tube songs  or. wikipedia songs .or google songs. . "
             "Would you like to give me some instructions? ")
        instruction = ListenToUser()
        if 's' in str(instruction) or 'yes' in str(instruction) or  'yeah' in str(instruction): 
                WebSearch(instruction)
        else:
            return
    except Exception as e:
        print(e)
        Speak("I didn't get you, Would you like to give me some instructions")
        instruction = ListenToUser()
        if 's' in str(instruction) or 'yes' in str(instruction) or  'yeah' in str(instruction):
            WebSearch(instruction)
            
```

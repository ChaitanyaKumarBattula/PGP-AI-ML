How to Build a feature Extraction Routine

Display how to build a labeled dataset from multiple input files.


```python
from nltk.tokenize.casual import EMOTICON_RE
```

```python
def age_features(message):
    emotion = EMOTICON_RE.findall(''.join(message)) 
    length = len(message)
    if not emotion:
        emotion = 0
    else:
        emotion = 1
    return{'Emotions':emotion, 'Length': length}

        
```

```python
from nltk.corpus import nps_chat

chatroom20 = nps_chat.posts('10-19-20s_706posts.xml')
chatroom30 = nps_chat.posts('10-19-30s_705posts.xml')
chatroom40 = nps_chat.posts('10-19-40s_686posts.xml')

#chatroom40[54]
```

```python
from nltk.classify import apply_features

train_set = apply_features(age_features, labeled_chats[500:])
test_set = apply_features(age_features, labeled_chats[:500])
```

```python
import nltk

classifier = nltk.NaiveBayesClassifier.train(train_set)
accuracy = (nltk.classify.accuracy(classifier, test_set))
accuracy
```

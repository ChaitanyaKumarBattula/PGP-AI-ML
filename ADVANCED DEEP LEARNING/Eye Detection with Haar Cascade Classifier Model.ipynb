```python
#Face Detection using Pre-trained model offered by OpenCV community

#We are going to use XML based model which is trained by OpenCV community
#Algorithm Called Viola Jones Algorithm
#Model : Haar Cascade Classifier Model


#Step1: Import Cv2

import cv2

#Step2: Load FrontalFace Model

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

#Step3: Create function to detect face

def detect_Face(img):
    
    imgCopy = img.copy()
    
    face_rect = face_cascade.detectMultiScale(imgCopy)
    
    for (x,y,w,h) in face_rect:
        cv2.rectangle(imgCopy, (x,y) , (x+w ,y+h), (0,255,0) , 5)

    return imgCopy

#Step4: Enable Camera
camera = cv2.VideoCapture(0)

#Step5: Show Camera Feed

while(True):
    _ ,frame = camera.read()
    
    frameWithFace = detect_Face(frame)
    
    #Display Frame
    cv2.imshow("My Camera Feed", frameWithFace)
    
    #Termination Condition
    if cv2.waitKey(1) & 0xFF ==  ord('q'):  #Press q to exit
        break
        
#Step4: Clean up Resources

camera.release()
cv2.destroyAllWindows()
```

```python
#Eye Detection using Pre-trained model offered by OpenCV community

#We are going to use XML based model which is trained by OpenCV community
#Algorithm Called Viola Jones Algorithm
#Model : Haar Cascade Classifier Model


#Step1: Import Cv2

import cv2

#Step2: Load FrontalEye Model

eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

#Step3: Create function to detect face

def detect_Face_Eye(img):
    
    imgCopy = img.copy()
    
    face_rect = face_cascade.detectMultiScale(imgCopy)
    
    
    for (x,y,w,h) in face_rect:
        cv2.rectangle(imgCopy, (x,y) , (x+w ,y+h), (0,255,0) , 5)
        
        eye_rect = eye_cascade.detectMultiScale(imgCopy)
        
        for (ex,ey,ew,eh) in eye_rect:
            cv2.rectangle(imgCopy, (ex,ey) , (ex+ew ,ey+eh), (255,0,0) , 5)

    return imgCopy

#Step4: Enable Camera
camera = cv2.VideoCapture(0)

#Step5: Show Camera Feed

while(True):
    _ ,frame = camera.read()
    
    frameWithFace = detect_Face_Eye(frame)
    
    #Display Frame
    cv2.imshow("My Camera Feed", frameWithFace)
    
    #Termination Condition
    if cv2.waitKey(1) & 0xFF ==  ord('q'):  #Press q to exit
        break
        
#Step4: Clean up Resources

camera.release()
cv2.destroyAllWindows()
```

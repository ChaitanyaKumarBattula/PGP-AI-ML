## EDA.1- Which movies have maximum views/ratings? 

We apparoch this from 2 different ways.
-1. Which movies have been rated by most viewers?
-2. Which movies have the aggregated high ratings?

Approach.1- Which movies have been rated by most viewers?

```python
df_main = df.copy()
df.describe().T["count"].sort_values(ascending = False)[:5].to_frame()
```

Approach2. Which movies have the high aggregated ratings?

```python
df.drop('user_id', axis=1).sum().sort_values(ascending=False)[:5].to_frame()
```

Conclusion:
We can see that, results obtained form both approaches is consistent.

## EDA.2(a)-  What is the average rating for each movie?

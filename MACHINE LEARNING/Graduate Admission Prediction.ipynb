# **TRAINING AND TESTING DATA**

```python
x = data.drop('Chance of Admit ',axis=1)
y = data['Chance of Admit ']
```

```python
from sklearn.linear_model import LassoLars,LinearRegression,LogisticRegression,Ridge,Lasso
from sklearn.metrics import r2_score,accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# Splitting training and testing data
x_train, x_test, y_train, y_test = train_test_split(x, y,
                     test_size = 0.3, random_state = 30)
 
```

```python
print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

# MODELS

**This is a Regression Problem so we will use regression models.**

# **LINEAR REGRESSION**

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
y_pred = lin_reg.predict(x_test)
 
print(f"r2 Score of test set : {r2_score(y_test, y_pred)}")
```

```python
lin_reg.intercept_
```

```python
lin_reg.coef_
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
model = LinearRegression(normalize=True)
model.fit(x_test, y_test)
model.score(x_test, y_test)*100
```

# **RIDGE REGRESSION**


```python
rid = Ridge(alpha=1.0)
rid.fit(x_train, y_train)
Ridge()
```

```python
y_pred = rid.predict(x_test)
 
print(f"r2 Score of test set : {r2_score(y_test, y_pred)}")
```

```python
rid.coef_
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mean_squared_error(y_test,y_pred)
```

```python
rid.score(x_test, y_test)*100
```

# **LASSO REGRESSION**

```python
las = Lasso(alpha=1.0)
las.fit(x_train, y_train)
Lasso()
```

```python
y_pred = las.predict(x_test)
 
print(f"r2 Score of test set : {r2_score(y_test, y_pred)}")
```

```python
las.coef_
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
mean_squared_error(y_test,y_pred)
```
